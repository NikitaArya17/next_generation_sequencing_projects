```python
meta = pd.read_csv("Metadata.txt", sep="\t")
meta.drop([0], axis = 0, inplace = True)
meta.rename(columns = {'NAME': 'CellID', 'Cell_line' : 'CellLine', 'Pool_ID' : 'Pool', 'Cancer_type' : 'Indication'}, inplace = True)
meta
```

```python
counts_cellid = pd.read_csv('UMIcount_data.txt', sep='\t', nrows = 1, header = None)
counts_cellid = counts_cellid.transpose()
counts_cellid.drop([0], inplace = True)
counts_cellid
```

```python
counts = pd.read_csv('UMIcount_data.txt', sep = '\t', skiprows = 3, header = None, index_col = 0)
counts = counts.transpose()
counts
```

```python
counts.index = counts_cellid[0]
counts.index.name = None
a = counts.index.isin(meta['CellID'])
counts = counts[a]
counts
```

```python
meta = meta.set_index('CellID')
meta = meta.reindex(index = counts.index)
meta
```

```python
counts.index.equals(meta.index)
```

```python
adata = anndata.AnnData(X = scipy.sparse.csr_matrix(counts), obs = meta, var = counts.columns.to_frame())
del counts
adata.var.drop(columns = [0], inplace = True)
adata.var.index.name = None
adata
```

```python
adata.obs.columns
```

```python
adata.obs['Genes_expressed'] = adata.obs['Genes_expressed'].astype(int)
columns = [column for column in adata.obs.columns if column != 'Genes_expressed' and column != 'n_genes']

for column in columns:
  if '_score' in column:
    adata.obs[column] = adata.obs[column].astype(float)
  else:
    adata.obs[column] = adata.obs[column].astype(str)
```

```python
sc.pp.filter_genes(adata, min_cells = 10)
sc.pp.filter_cells(adata, min_genes = 200)
adata
```

```python
adata.write('anndata.h5ad')
```
